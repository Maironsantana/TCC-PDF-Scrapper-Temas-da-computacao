```python
#Insira o local do arquivo PDF
pdf_file = '/content/drive/MyDrive/...'

#Função irá realizar a contagem de frequência de palavras
#Ele usa o dicionário do tema para verificar a quantidade de vezes que uma determinada palavra apareceu
def contar_palavras_chave(pdf_file, palavras_chave):
    contador = 0

    with open(pdf_file, 'rb') as pdf:
        pdf_reader = PdfReader(pdf)
        total_paginas = len(pdf_reader.pages)

        for pagina_num in range(total_paginas):
            pagina = pdf_reader.pages[pagina_num]
            texto = pagina.extract_text()
            todas_palavras = texto.split()

            for palavra in palavras_chave:
                if palavra in texto:
                    contador += texto.count(palavra)
                    # print(palavra)
    return contador


# Para usar efetivamente, vamos fazer um laço de repetição no dicionário unificado e buscar para todos os temas
# Os valores de frequência estão sendo salvos na variável frequencia

frequencia = {}
for index, topico in dicionario_unificado.items():
  frequencia[index] = {}
  total_encontrado = contar_palavras_chave(pdf_file, topico)
  frequencia[index] = total_encontrado
  print("No tema", index,f" teve o total de palavras-chave encontradas: {total_encontrado}")

```

```python
#para descobrirmos mais facilmente a relação de um tema com o texto, é feito um calculo simples de porcentagem para cada tema

soma = 0
for valor in frequencia.values():
  soma += valor

porcentagens = {}

for index, valor in frequencia.items():
  porcentagens[index] = valor/soma

porcentagens
```

Ao final você terá para um arquivo PDF, a porcentagem de relação com cada tema.
